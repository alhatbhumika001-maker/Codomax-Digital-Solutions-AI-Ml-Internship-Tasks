```python
import numpy as np

hours = np.array([1,2,3,4,5,6,7,8])

print("Study Hours:", hours)
print("Mean Hours:", np.mean(hours))
print("Maximum Hours:", np.max(hours))
print("Minimum Hours:", np.min(hours))
print("Total Hours:", np.sum(hours))
```
