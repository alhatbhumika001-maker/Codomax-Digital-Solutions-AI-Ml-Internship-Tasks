```python
# Calculate Average Marks

math = 85
science = 90
english = 80

average = (math + science + english) / 3

print("Average Marks:", average)
```

```python
# Function

def greet(name):
    print("Welcome", name)

greet("Bhumika")
```

```python
# While Loop

count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
# For Loop

print("Numbers from 1 to 5")

for i in range(1, 6):
    print(i)
```

```python
# If Else

marks = 75

if marks >= 40:
    print("Result: Pass")
else:
    print("Result: Fail")
```

```python
# Arithmetic Operators

a = 20
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
```

```python
# Data Types

integer_num = 10
float_num = 12.5
text = "Machine Learning"
boolean = True

print(type(integer_num))
print(type(float_num))
print(type(text))
print(type(boolean))
```
